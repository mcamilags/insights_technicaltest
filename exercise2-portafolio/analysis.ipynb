```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import os
```

```python
# Leer los datos desde el archivo Excel
file = "../files/prueba.xlsx"

R_raw = pd.read_excel(file, sheet_name="Matriz de Simulacion", header=0)
P1_raw = pd.read_excel(file, sheet_name="P1", header=0)
P2_raw = pd.read_excel(file, sheet_name="P2", header=0)
```

## 1. Fórmulas en términos de P y R

Tenemos:
- **R** = matriz de retornos simulados [30,000 × 55]
- **P** = vector de pesos del portafolio [55]

Primero calculamos el retorno del portafolio en cada escenario:

$$r_p = R \cdot P$$

Esto nos da un vector de 30,000 retornos, uno por escenario.

**Retorno esperado** — promedio de esos 30,000 valores:

$$E[r_p] = \frac{1}{N} \sum_{i=1}^{N} r_{p,i}$$

**Volatilidad esperada** — qué tan dispersos están esos valores alrededor del promedio:

$$\sigma_p = \sqrt{\frac{1}{N} \sum_{i=1}^{N} \left(r_{p,i} - E[r_p]\right)^2}$$


## 2. Cálculo numérico

```python
# Convertir los datos a arrays numéricos 
R  = R_raw.values.astype(float)           # [30000 x 55]
P1 = P1_raw.iloc[0].values.astype(float)  # [55]
P2 = P2_raw.iloc[0].values.astype(float)  # [55]
```

```python
# Retornos de portafolio
Rp1 = R @ P1
Rp2 = R @ P2

# Retorno esperado
ret_esp_p1 = np.mean(Rp1)
ret_esp_p2 = np.mean(Rp2)

# Volatilidad
vol_p1 = np.std(Rp1)
vol_p2 = np.std(Rp2)

print("Portafolio 1")
print("Retorno esperado:", ret_esp_p1)
print("Volatilidad:", vol_p1)

print("Portafolio 2")
print("Retorno esperado:", ret_esp_p2)
print("Volatilidad:", vol_p2)
```

### Claude
¿Cómo afecta una subida de tasas de la Fed al retorno y volatilidad esperados de un portafolio de renta fija y uno de renta variable?
Cuando la Fed sube tasas, el efecto es distinto dependiendo del tipo de portafolio:

**Renta Fija (bonos):**

Los bonos existentes pierden valor porque los nuevos bonos que se emiten pagan tasas más altas — nadie quiere los viejos. Esta relación inversa entre tasas y precio es matemáticamente directa.
El retorno esperado cae en el corto plazo (caída de precios), pero sube en el largo plazo porque los nuevos bonos que compras rinden más.
La volatilidad aumenta, especialmente en bonos de largo plazo que son más sensibles a cambios de tasas (mayor duración).

**Renta Variable (acciones):**

Las empresas pagan más por su deuda → menores utilidades → valuaciones más bajas.
Los flujos futuros se descuentan a una tasa mayor → el valor presente de las acciones cae.
El retorno esperado se comprime en el corto plazo.
La volatilidad aumenta porque hay más incertidumbre sobre crecimiento económico y hay rotación de capital desde acciones hacia bonos (que ahora son más atractivos).


**En resumen:** una subida de tasas es generalmente negativa para ambos portafolios en el corto plazo, pero el impacto es más predecible y mecánico en renta fija, mientras que en renta variable depende más del contexto económico.

## 1. Renta Fija (Bonos)

Es el activo que siente el impacto de forma más directa y matemática.

* **Retorno Esperado:**
* **Corto Plazo:** El precio de los bonos existentes **cae**. Existe una relación inversa: si las tasas nuevas son más altas, tu bono viejo (que paga menos) pierde atractivo y su precio baja para compensar.
* **Largo Plazo:** El retorno esperado **sube**. A medida que los bonos vencen, puedes reinvertir ese capital en nuevos papeles con cupones más jugosos.


* **Volatilidad:**
* **Aumenta**, especialmente en bonos de larga duración. Los inversores intentan adivinar hasta dónde llegará la Fed, lo que provoca oscilaciones bruscas en los precios de mercado.



---

## 2. Renta Variable (Acciones)

Aquí el efecto es más psicológico y estructural, basado en la valoración de las empresas.

* **Retorno Esperado:**
* **Generalmente disminuye o se vuelve más exigente.** Al subir las tasas, la "tasa de descuento" que usamos para valorar las ganancias futuras de una empresa aumenta. Matemáticamente, esto hace que el valor presente de la empresa sea menor.
* **Costos financieros:** A las empresas les cuesta más pedir prestado, lo que puede reducir sus márgenes de beneficio y frenar el crecimiento.


* **Volatilidad:**
* **Se dispara.** El mercado odia la incertidumbre. Las acciones de "crecimiento" (como las tecnológicas) suelen ser las más golpeadas porque su valor depende de flujos de caja muy lejanos en el tiempo, los cuales se ven más castigados por el aumento de tasas.
